```python
import pandas as pd
import sqlite3
```

```python
connection=sqlite3.connect("books.db")
```

```python
query=lambda x,**kwargs : pd.read_sql(x,connection,**kwargs)
```

# A
Select all authors last names from the authors table in descending order

```python
query("""SELECT last FROM authors ORDER BY last DESC""")
```

# B
Select all book titles from the title tables in ascending order

```python
query("""SELECT title FROM titles ORDER BY title ASC""")
```

# C
Use an INNER Join to select all books for a specific author.

```python
query("""SELECT title, copyright, titles.isbn FROM titles, author_ISBN INNER JOIN authors ON author_ISBN.id = authors.id WHERE authors.id = 3 ORDER BY title ASC""") 
```
